```python
import pandas as pd
df = pd.read_csv('advertising.csv')
df.head(20)
```

```python
x = df.drop(['Ad Topic Line','City','Country','Clicked on Ad','Timestamp'],axis=1)
y = df['Clicked on Ad']
x.head()
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_reg = Lasso(alpha=0.1)
lasso_reg.fit(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=5)
```

```python
lasso_reg.score(x_test,y_test)
```

```python
lasso_reg.score(x_train,y_train)
```

```python
import numpy as np
coeff_used = np.sum(lasso_reg.coef_!=0)
print(coeff_used)
print(df.columns[[0,1,2,3,6]][lasso_reg.coef_!=0])
```

```python
x_new = df[df.columns[[0,1,2,3]]]
lasso_reg.fit(x_new,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_new,y,test_size=0.3,random_state=5)
```

```python
lasso_reg.score(x_test,y_test)
```

```python
lasso_reg.score(x_train,y_train)
```
